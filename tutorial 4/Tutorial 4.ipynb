$$ \mathbf{k}_1 = h\,\mathbf{f}(t_n,\mathbf{y}_n), $$ 

$$ \mathbf{k}_2 = h\,\mathbf{f}\Big(t_n+\frac{h}{2},\mathbf{y}_n+\frac{\mathbf{k}_1}{2}\Big), $$

$$ \mathbf{y}_{n+1} \approx \mathbf{y}_n+\mathbf{k}_2, $$

where $\mathbf{y}$ is the vector containing solutions to the ODEs that describe your problem and $\mathbf{f}$ is the derivative function (see paragraph 8.4 for exact definitions of these).
